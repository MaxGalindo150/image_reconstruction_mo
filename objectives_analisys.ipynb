```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results_df = pd.read_csv("solutions_comparison.csv")
```

```python
results_df
```

```python
correlation_matrix = results_df[["f1", "f2", "f3"]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Matiz de correlación entre las funciones objetivo")
plt.show()
```

```python
objectives = results_df[['f1', 'f2', 'f3']]

pairplot = sns.pairplot(objectives, diag_kind="kde", markers="o", plot_kws={'alpha': 0.6})

pairplot.fig.suptitle("Diagramas de Dispersión por Pares de Objetivos", y=1.02, fontsize=16)
pairplot.fig.tight_layout()

plt.savefig("pairwise_objectives.png", dpi=300, bbox_inches="tight")
plt.show()
```
