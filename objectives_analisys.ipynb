```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results_df = pd.read_csv("results_pareto_single_value.csv")
```

```python
results_df
```

```python
correlation_matrix = results_df[["f1", "f2", "f3"]].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Correlation Matrix of Objectives")
plt.show()
```

```python
objectives = results_df[['f1', 'f2', 'f3']]

sns.pairplot(objectives, diag_kind="kde", markers="o", plot_kws={'alpha': 0.6})
plt.suptitle("Pairwise Scatterplots of Objectives", y=1.02, fontsize=16)
plt.savefig("pairwise_objectives.png")
plt.show()
```
